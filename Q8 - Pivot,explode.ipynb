```python
from pyspark.sql.functions import arrays_zip, explode, split, col, lit, expr
```

```python
l = [(1,'Gaga','India',"2022-01-11"),(1,'Katy','UK', "2022-01-11"),(1,'Bey','Europe', "2022-01-11"),(2,'Gaga',None,"2022-10-11"),
     (2,'Katy','India', "2022-10-11"),(2,'Bey','US',"2022-02-15"),(3,'Gaga','Europe', "2022-10-11"),
     (3,'Katy','US',"2022-10-11"),(3,'Bey',None,"2022-02-15"), (1, 'Gaga', 'US',"2022-01-11"),(3, 'Katy', 'Switz',"2022-02-15") ]

df = spark.createDataFrame(l, ['ID','NAME','COUNTRY', 'Date_part'])
display(df)
```

```python
df1 =  df.groupBy("ID", "Date_part").pivot("NAME").agg(collect_list("COUNTRY"))
df1.display()
```

```python
df2= df1.withColumn("zipped",arrays_zip(*df1.columns[2:]))
df2.display()
```

```python
df3 =  df2.withColumn("exploded", explode("zipped"))
df3.display()
```

```python
df4 =  df3.select("ID","DATE_part", "exploded.*")
df4.display()   
```

```python
from pyspark.sql.functions import collect_list
df2 =  df.groupBy("ID", "DATE_part").pivot("NAME").agg(collect_list("COUNTRY"))

df2.display()
```

```python
df3= df2.withColumn("arrays_zipped", arrays_zip("Bey", "Gaga", "Katy"))
df3.display()

```

```python

df4=  df3.withColumn("exploded",explode("arrays_zipped"))
df4.display()
```

```python
from pyspark.sql.functions import expr
df5 =  df4.select("ID","DATE_part", "exploded.*")

df5.display()
```
