```python
import regex as re
from pyspark.sql.functions import *
```

```python
df  = spark.read.text('/Volumes/study/geekcoders/raw_data/Regex.txt')
df.display()
```

```python
df1= df.withColumn("value_split", regexp_replace(col("value"), r"([A-Z])-(\d+)-", r'$1-$2,'))
df1.display()
```

```python
df2=df1.withColumn("split_value", split(col("value_split"), ",")).drop("value").drop("value_split")
df2.display()
```

```python
df3= df2.withColumn("exploded", explode(col("split_value"))).withColumn("id", split(col("exploded"), "-").getItem(0))\
.withColumn("name", split(col("exploded"), "-").getItem(1)).withColumn("age", split(col("exploded"), "-").getItem(2))
df3.display()
```
