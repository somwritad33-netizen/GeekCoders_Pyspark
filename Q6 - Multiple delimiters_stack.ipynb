```python
df = spark.read.csv("/Volumes/study/geekcoders/raw_data/Multiplt_delimiters.csv", header=True, inferSchema=True)
df.display()
```

```python
from pyspark.sql.functions import split, col
df1 = df.withColumn("Maths", split(col("Marks"), "\\|")[0])\
    .withColumn("Science", split(col("Marks"), "\\|")[1])\
    .withColumn("English", split(col("Marks"), "\\|")[2])\
    .withColumn("History", split(col("Marks"), "\\|")[3])\
    .drop("Marks")
  

df1.display()
```

```python
df2 = df1.selectExpr(
    "ID", "NAME", "AGE",
    "stack(4, 'Maths', Maths, 'Science', Science, 'English', English, 'History', History) as (Subject, Marks)"
)

df2.show()
```
