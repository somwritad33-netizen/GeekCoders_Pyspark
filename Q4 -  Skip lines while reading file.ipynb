```python
from pyspark.sql.functions import lit, row_number, col, split
from pyspark.sql.window import Window
```

```python
df = spark.read.csv("/Volumes/study/geekcoders/raw_data/Sample.csv", header=True, inferSchema=True)
df.display()
```

```python
from pyspark.sql.functions import monotonically_increasing_id
# Add an index column to the DataFrame
df_with_index = df.withColumn('index', monotonically_increasing_id())
df_with_index.display()
```

```python
#Convert index_correct to header

index_correct= df_with_index.filter(col("_c1") == 'NAME').select("index").first()[0]
print(index_correct)
header  =  df_with_index.filter(col("index")==f'{index_correct}')
header_values =(header.drop("index").collect()[0])
l=[]
l.append((header_values[0], header_values[1]))
print(l)
```

```python
#Take actual data 

data  =  df_with_index.filter(col("index")>f'{index_correct}')
data_values =list(data.drop("index").collect())
print(data_values)

for row in data_values:
    l.append((row[0], row[1]))
print(l)
```

```python
df_final =spark.createDataFrame(l[1:], l[0])
display(df_final)
```
