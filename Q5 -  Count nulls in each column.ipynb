```python
from pyspark.sql.functions import col, sum, count, when
```

```python
df = spark.read.csv("/Volumes/study/geekcoders/raw_data/Count_nulls_in_each_column.csv", header=True, inferSchema=True)
df.display()
```

```python
#print column names in the dataframe

cols=df.columns
print(cols)
```

```python
#iterate over columns
for colname in cols:
 row1 = df.select(col(f'{colname}')).collect()
 l=[]
 print(row1)
 for r in row1:
     l.append(r[0]) 
 print(colname, l.count("null"))
```
