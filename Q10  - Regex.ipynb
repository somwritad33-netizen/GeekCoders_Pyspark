```python
import regex as re
from pyspark.sql.functions import *
```

```python
data=[('ABSHFJFJ12QWERT12',1),('QWERT5674OTUT1',2),('DGDGNJDJ1234UYI',3)]
df=spark.createDataFrame(data,schema="input_string string,id int")
df.show()
```

```python
df1 = df.withColumn("fpart", regexp_replace("input_string", r"([A-Z][0-9]+).*", r"$1"))
df2 =  df1.withColumn("lpart", replace("input_string", df["fpart"], lit("")))
df3  =  df2.withColumn("lpart1", split(regexp_replace("input_string",r"([A-Z][0-9]+)([A-Z0-9]+)", r"$1,$2"), ",")[1])
df3.show()
                                      
```
